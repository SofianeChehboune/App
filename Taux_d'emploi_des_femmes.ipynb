## 5/ *Creation of the animated graphic with Plotly for 500 first rows (https://plotly.com/)*

```python
fig = px.bar(df.head(500),
             x="Taux d'emploi des femmes (%)",
             y="Pays",
             color="Pays",
             animation_frame="AnnÃ©e",
             orientation="h",
             title="Ã‰volution du taux d'emploi des femmes (1991-2023)",
             range_x=[0, 100],  # Ajuste l'Ã©chelle si nÃ©cessaire
             width=1400,   # Largeur en pixels
             height=600)   # Hauteur en pixels

fig.show()
```
